```python
import smtplib     # Server and Application layer protocols
import ssl         # Secured Encryption
import getpass     # hides the password while typing
from email import encoders
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

import openpyxl
```

```python
excel = './data/passwords.xlsx'
obj = openpyxl.load_workbook(excel) 
sheet = obj.active 
```

```python
sender_email = 'acm@gmrit.edu.in'

password = getpass.getpass(prompt="Enter Password", stream=None)

port = 587        # SSL PORT
smtp_server = "smtp-mail.outlook.com"
```

```python
count=508
while (True):
    # get EXCEL data
    get_name = sheet.cell(row = count ,column = 1)
    get_email = sheet.cell(row = count, column = 2)
    get_password = sheet.cell(row = count ,column = 3)
    name = get_name.value
    user_password = get_password.value
    user_email = get_email.value
    jntu = user_email.split('@')[0]
    if(name == None):
        break
    name = name.upper()
    
    
#     Working Area

    subject = "VI Semester Open Elective Password"

    text = f"""\
Subject: VI Semester Open Elective Password

Hello {name},
Below are the username and password for your Open Elective Registration Process
Username: JNTU Number
Password: {user_password}\
    """

    html = """\
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta http-equiv="X-UA-Compatible" content="IE=edge">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Open Elective Registration</title>

    <style>
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }

        body {
            background-color: rgba(255, 248, 220, 0.3);
        }

        #main {
            background-color: #fff;
            max-width: 720px;
            margin: auto;
            border-left: 1px solid #ccc;
            border-right: 1px solid #ccc;
            border-bottom: 5px solid #013b8f;
        }

        .header {
            background-color: #013b8f;
            height: 70px;
            width: 100%;
            border-radius: 5px;
            display: flex;
            align-items: center;
            justify-content: center;
        }

        .container {
            padding: 1.25rem;
            max-width: 100%;
        }

        #description {
            text-align: justify;
            margin-top: 1.5rem;
            line-height: 1.3rem;
        }

        #info {
            text-align: justify;
        }

        table {
            margin: auto;
            margin-top: 20px;
        }

        table,
        th,
        td {
            border-collapse: collapse;
            border: 1px solid black;
            padding: 10px;
        }
    </style>
</head>
"""
    html += f"""\
<body>
    <div id="main">
        <div class="header">
            <p style="color: #fff; font-size: 1.2rem; text-align: center;">VI Semester Open Elective Registration</p>
        </div>
        <div class="container">
            <section id="info">
                <p id="description">
                    Hello {name},<br>
                    Below are the username and password for your Open Elective Registration Process
                    <h1 style="text-align: center; margin-top: 20px">Username: {jntu}</h1>
                    <h1 style="text-align: center; margin-top: 20px">Password: {user_password}</h1>
                </p>
                <br>
                <p>Note: Please check the credentials prior to the registration start time. For any queries reach out to acm@gmrit.edu.in</p>
            </section>
        </div>
    </div>
</body>

</html>\
    """
    
    receiver_email = user_email
    
    
    message = MIMEMultipart("alternative")
    message["Subject"] = subject
    message["From"] = sender_email
    message["To"] = receiver_email


    plain_body = MIMEText(text, "plain")
    html_body = MIMEText(html, "html")

    message.attach(plain_body)
    message.attach(html_body)


    with smtplib.SMTP(smtp_server, port) as server:
        try:
            server.starttls(context=ssl.create_default_context())
            server.login(sender_email, password)
        except Exception as e:
            print(e, "or Login Error")
        server.sendmail(sender_email, receiver_email, message.as_string())


    print(count, name, user_email, user_password)
#     if count == 3:
#         break
    count+=1
    
print('Exited...')
```
